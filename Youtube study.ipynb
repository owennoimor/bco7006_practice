### **Condition**
if, elif, else


```python
score = 79.9

if score >= 80:
  print('Grade A')

elif score >= 70:
  print('Grade B')

elif score >= 60:
  print('Grade C')

elif score >= 50:
  print('Grade D')

else:
  print('Grade F')
```

```python
for a in range(1, 7):
  if a % 2 == 0:
    continue # continue = skip loop
  print(a)
```

```python
for a in range(1, 7):
  if a % 2 == 0:
    break # break = stop loop
  print(a)
```

### Function
def function():

in side () after function = parameter

```python
def get_box_area():
  w = 4
  l = 4
  h = 2
  box_area = w * l * h
  print(box_area)

get_box_area()
```

```python
def get_box_area(w, l, h):
  box_area = w * l * h
  print(box_area)

get_box_area(4, 4, 2) # can put it directly
get_box_area(w=1, l=2, h=3) # can define each parameter as well
```

```python
def get_box_area(w, l, h):
  box_area = w * l * h
  return box_area              # use return เพื่อคืนค่ากลับไปแบบไม่แสดงออกมา

box1 = get_box_area(4, 4, 2)   # เลยมา set new variable
box2 = get_box_area(w=1, l=2, h=3)

print(box1)                    # print later
```

```python
def get_box_area(w, l, h):
  if w < 0 or l < 0 or h < 0:
    return 0                      # set some condition first
  box_area = w * l * h
  return box_area                 # if pass then return function

box1 = get_box_area(4, -4, 2)
box2 = get_box_area(w=1, l=2, h=3)

print(box1)
```

### Dictionary

```python
book = {'name': 'C++', 'price': 500, 'page': 100}

print(book)
```

```python
book = {
  'name': 'C++',
  'price': 500,
  'page': 100
}

print(book['name']) # want to print specific thing, has to use []
```

```python
book = {
  'name': 'C++',
  'price': 500,
  'page': 100
}
book['price'] = 400 # replace has to use []
book['publisher'] = 'Melbourne' # if the replacement is not match, it will add new one

print(book)
```
